```python
# Using: Python 3.7.6, NetworkX 2.4
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
plt.style.use('seaborn-white')
plt.rcParams["font.size"] = 20.0
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams['figure.dpi'] = 300
```

```python
K5 = nx.complete_graph(5)
```

```python
nx.draw(K5, with_labels=True, node_size=2000)
```

```python
random.choice([n for n in K5.neighbors(0)])
```

```python
walk = [0]
for i in range(0,10000):
    walk.append(random.choice([n for n in K5.neighbors(walk[i])]))
```

```python
plt.hist(walk, bins=5, rwidth=0.9)
plt.xlabel("Vertex labels")
plt.ylabel("Number of vertex vistiations in a random walk of length $10^4$")
plt.xticks([0,1,2,3,4], ha='right')
plt.show()
```

```python
K5_K5 = nx.union(K5, K5, rename=('0-', '1-'))
K5_K5.add_edge('0-0', '1-0')
```

```python
nx.draw(K5_K5, with_labels=True, node_size=2000)
```

```python
walk2 = ['0-0']
for i in range(0,100_000):
    walk2.append(random.choice([n for n in K5_K5.neighbors(walk2[i])]))
```

```python
plt.hist(walk2, bins=10, rwidth=0.9)
plt.xlabel("Vertex labels")
plt.ylabel("Number of vertex vistiations in a random walk of length $10^4$")
plt.xticks(rotation=45)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.show()
```

```python
K5_K5_K5 = nx.union(K5_K5, K5, rename=('0-', '1-'))
K5_K5_K5.add_edge('0-0-1', '1-0')
```

```python
nx.draw(K5_K5_K5, with_labels=True, node_size=1500)
```

```python
walk3 = ['0-0-0']
for i in range(0,100_000_000):
    walk3.append(random.choice([n for n in K5_K5_K5.neighbors(walk3[i])]))
```

```python
plt.hist(walk3, bins=15, rwidth=0.9)
plt.xticks(rotation=45)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.xlabel("Vertex labels")
plt.ylabel("Number of vertex vistiations in a random walk of length $10^8$")
plt.show()
```
